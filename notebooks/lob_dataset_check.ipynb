Inspect X/y and missing ratios for LOBDS.

```python
from pathlib import Path
import numpy as np
import pandas as pd

# Change date if needed
trade_day = "20260105"

clean_root = Path(r"D:/cbond_on/clean_data")
ds_dir = clean_root / "LOBDS" / trade_day
ds_dir
```

```python
X = np.load(ds_dir / "X.npy")
y = np.load(ds_dir / "y.npy")
meta = pd.read_parquet(ds_dir / "meta.parquet")

print("X shape:", X.shape)
print("y shape:", y.shape)
meta.head(3)
```

```python
# NaN ratio in X and y
nan_ratio_X = np.isnan(X).mean()
nan_ratio_y = np.isnan(y).mean()
print("NaN ratio X:", nan_ratio_X)
print("NaN ratio y:", nan_ratio_y)
```

```python
# per-sample NaN ratio (should be 0 if ffill worked)
nan_ratio_per_sample = np.isnan(X).mean(axis=(1, 2, 3))
pd.Series(nan_ratio_per_sample).describe()
```
