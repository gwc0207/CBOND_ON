```python
from __future__ import annotations

from pathlib import Path
from datetime import datetime
import sys

# ???????????? cbond_on ????
_CANDIDATES = [Path.cwd(), Path.cwd().parent, Path.cwd().parent.parent]
PROJECT_ROOT = None
for _p in _CANDIDATES:
    if (_p / "cbond_on").exists() and (_p / "cbond_on").is_dir():
        PROJECT_ROOT = _p
        break
if PROJECT_ROOT is None:
    raise RuntimeError("Cannot locate project root containing 'cbond_on' package directory")
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import numpy as np
import pandas as pd

from cbond_on.core.config import load_config_file, parse_date
from cbond_on.infra.data.panel import read_panel_data
from cbond_on.app.usecases.factor_batch_runtime import build_signal_specs
from cbond_on.domain.factors import defs  # noqa: F401
from cbond_on.domain.factors.builder import build_factor_frame
from cbond_on.infra.factors.rust_backend import build_factor_frame_rust
from cbond_on.infra.factors.rust_shm_backend import build_factor_frame_rust_shm

print(f"PROJECT_ROOT={PROJECT_ROOT}")
```

```python
# ---- ???? ----
DAY = "2025-01-03"
ENGINE_A = "python"      # python | rust | rust_shm_exp
ENGINE_B = "rust"        # python | rust | rust_shm_exp
FACTOR_WORKERS = 2        # rust_shm_exp ??
TOPK = 20

# ????????????None ????
ONLY_FACTORS: list[str] | None = None
```

```python
def load_panel_and_specs(day_text: str):
    cfg = load_config_file("factor")
    paths = load_config_file("paths")
    day = parse_date(day_text)
    panel_name = str(cfg.get("panel_name", "")).strip()
    if not panel_name:
        raise ValueError("factor_config.panel_name is required")

    panel = read_panel_data(
        paths["panel_data_root"],
        day,
        window_minutes=15,
        panel_name=panel_name,
    ).data
    if panel is None or panel.empty:
        raise RuntimeError(f"panel is empty for day={day} panel_name={panel_name}")

    specs = build_signal_specs(cfg)
    if ONLY_FACTORS:
        allow = {x.strip() for x in ONLY_FACTORS if str(x).strip()}
        specs = [s for s in specs if s.name in allow or (s.output_col or s.name) in allow]
    if not specs:
        raise RuntimeError("no specs selected")

    return panel, specs, panel_name


def compute_by_engine(panel: pd.DataFrame, specs, engine: str, factor_workers: int = 2) -> pd.DataFrame:
    engine = str(engine).strip().lower()
    if engine == "python":
        return build_factor_frame(panel, specs, workers=1, compute_backend_params={})

    if engine == "rust":
        return build_factor_frame_rust(panel, specs, compute_backend_params={})

    if engine == "rust_shm_exp":
        compute_params = {
            "__compute_backend__": {
                "shm_mp_start_method": "auto",
                "shm_fallback_to_rust": True,
                "shm_fork_zero_copy": True,
            }
        }
        return build_factor_frame_rust_shm(
            panel,
            specs,
            compute_backend_params=compute_params,
            workers=max(1, int(factor_workers)),
        )

    raise ValueError(f"unsupported engine: {engine}")


def align_frames(df_a: pd.DataFrame, df_b: pd.DataFrame):
    common_cols = [c for c in df_a.columns if c in df_b.columns]
    if not common_cols:
        raise RuntimeError("no common factor columns between two engines")

    left = df_a[common_cols].sort_index()
    right = df_b[common_cols].sort_index()
    common_idx = left.index.intersection(right.index)
    if len(common_idx) == 0:
        raise RuntimeError("no common index rows between two outputs")

    return left.loc[common_idx], right.loc[common_idx]


def build_parity_summary(df_a: pd.DataFrame, df_b: pd.DataFrame) -> pd.DataFrame:
    left, right = align_frames(df_a, df_b)
    rows = []
    for col in left.columns:
        a = pd.to_numeric(left[col], errors="coerce")
        b = pd.to_numeric(right[col], errors="coerce")
        mask = a.notna() & b.notna()
        n_valid = int(mask.sum())

        if n_valid == 0:
            rows.append({
                "factor": col,
                "n_aligned": int(len(left)),
                "n_valid": 0,
                "max_abs": np.nan,
                "mae": np.nan,
                "rmse": np.nan,
                "p99_abs": np.nan,
                "match_tol_1e_12": np.nan,
                "match_tol_1e_9": np.nan,
                "match_tol_1e_6": np.nan,
            })
            continue

        diff = (a[mask] - b[mask]).abs().astype(float)
        rows.append({
            "factor": col,
            "n_aligned": int(len(left)),
            "n_valid": n_valid,
            "max_abs": float(diff.max()),
            "mae": float(diff.mean()),
            "rmse": float(np.sqrt((diff ** 2).mean())),
            "p99_abs": float(diff.quantile(0.99)),
            "match_tol_1e_12": float((diff <= 1e-12).mean()),
            "match_tol_1e_9": float((diff <= 1e-9).mean()),
            "match_tol_1e_6": float((diff <= 1e-6).mean()),
        })

    out = pd.DataFrame(rows)
    return out.sort_values(["max_abs", "mae"], ascending=[False, False]).reset_index(drop=True)


def factor_top_diff(df_a: pd.DataFrame, df_b: pd.DataFrame, factor: str, topn: int = 20) -> pd.DataFrame:
    left, right = align_frames(df_a, df_b)
    if factor not in left.columns:
        raise KeyError(f"factor not found: {factor}")

    a = pd.to_numeric(left[factor], errors="coerce")
    b = pd.to_numeric(right[factor], errors="coerce")
    mask = a.notna() & b.notna()

    detail = pd.DataFrame({
        "algo_a": a[mask],
        "algo_b": b[mask],
    })
    detail["abs_diff"] = (detail["algo_a"] - detail["algo_b"]).abs()
    return detail.sort_values("abs_diff", ascending=False).head(int(topn))
```

```python
panel, specs, panel_name = load_panel_and_specs(DAY)
print(f"day={DAY} panel_name={panel_name} panel_rows={len(panel)} specs={len(specs)}")

df_a = compute_by_engine(panel, specs, ENGINE_A, factor_workers=FACTOR_WORKERS)
df_b = compute_by_engine(panel, specs, ENGINE_B, factor_workers=FACTOR_WORKERS)

summary = build_parity_summary(df_a, df_b)
summary.head(TOPK)
```

```python
# ???????????
worst_factor = str(summary.iloc[0]["factor"])
print("worst_factor:", worst_factor)
factor_top_diff(df_a, df_b, worst_factor, topn=30)
```

```python
# ????
out_dir = Path("runtime/diagnostics/factor_parity") / datetime.now().strftime("%Y%m%d_%H%M%S")
out_dir.mkdir(parents=True, exist_ok=True)

summary.to_csv(out_dir / "parity_summary.csv", index=False)
factor_top_diff(df_a, df_b, str(summary.iloc[0]["factor"]), topn=200).to_csv(
    out_dir / "worst_factor_top_diff.csv"
)

meta = {
    "day": DAY,
    "panel_name": panel_name,
    "engine_a": ENGINE_A,
    "engine_b": ENGINE_B,
    "factor_workers": int(FACTOR_WORKERS),
    "n_specs": int(len(specs)),
    "n_common_cols": int(len(summary)),
    "global_max_abs": float(summary["max_abs"].max()),
    "out_dir": str(out_dir.resolve()),
}
(out_dir / "meta.json").write_text(pd.Series(meta).to_json(indent=2), encoding="utf-8")
print("saved:", out_dir.resolve())
```
